```python
from causallearn.search.Granger.Granger import Granger
G = Granger(maxlag = 3)
# p_value_matrix = G.granger_test_2d(df)
coeff = G.granger_lasso(df.values)
```

```python
coeff
```

```python
# coeff: coefficient matrix [A_1, A_2, ..], where A_k is the dxd causal matrix for the k-th time lag. The ij-th entry
#         in A_k represents the causal influence from j-th variable to the i-th variable.
```

```python
plt.figure(figsize=(10,8)) # you can adjust the figure size as per your requirements
sns.heatmap(coeff[:coeff.shape[0], coeff.shape[0]:2*coeff.shape[0]], cmap='viridis')
plt.show()
```
